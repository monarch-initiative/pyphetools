```python
import pyphetools
import pandas as pd
```

# Extract the HPO seed terms separately for individuals and combine them into a cohort

```python
hpo_parser = pyphetools.creation.HpoParser(hpo_json_file="../hp.json")
```

```python
hpo_cr = hpo_parser.get_hpo_concept_recognizer()
# ontology = hpo_parser.get_ontology()
```

```python
def get_hpo_terms_from_text(text_input_curator: str):
    """
    Extract and return HPO terms from an input text
    """
    hpo_terms = set()
    for line in text_input_curator.split("\n"):
        hpo_term_list = hpo_cr.parse_cell(line)
        for hpot in hpo_term_list:
            hpo_terms.add(hpot)
    return hpo_terms
```

```python
class Individual_ET():

    def __init__(self, id, pmid, pmid_title, disease_id, disease_label,
                 hgnc_id, gene_symbol, transcript_id, allele_1, allele_2, variant_comment,
                 age_of_onset, age_at_last_encounter, deceased,
                 sex, hpo, text_input_curator, comment='na'):
        self.id = id # Higgins-Patient-1
        self.pmid = pmid # PMID:28289718
        self.pmid_title = pmid_title # Elucidation of MRAS-mediated Noonan syndrome with cardiac hypertrophy
        self.disease_id = disease_id # OMIM:618499
        self.disease_label = disease_label # Noonan syndrome-11
        self.hgnc_id = hgnc_id # HGNC:7227
        self.gene_symbol = gene_symbol # MRAS
        self.transcript_id = transcript_id # NM_001085049.3
        self.allele_1 = allele_1 # c.68G>T
        self.allele_2 = allele_2 # na
        self.variant_comment = variant_comment # Original notation NM_012219, c.68G>T
        self.age_of_onset = age_of_onset # P1Y8M0D
        self.age_at_last_encounter = age_at_last_encounter # P17Y
        self.deceased = deceased # yes|no|na
        self.sex = sex # M|F|O|U
        self.hpo = hpo # na
        self.text_input_curator = text_input_curator
        self.hpo_terms = get_hpo_terms_from_text(text_input_curator) # Set of ontolgy terms, e.g. (HP:0001714, "Ventricular hypertrophy")
        self.comment = None # na

    def get_pandas_df(self):
        # Set base columns
        columns = ["PMID", "title", "individual_id", "comment", "disease_id", "disease_label"]
        columns += ["HGNC_id", "gene_symbol", "transcript", "allele_1", "allele_2", "variant.comment"]
        columns += ["age_of_onset", "age_at_last_encounter", "deceased", "sex", "HPO"]
        # Initialize dataframe
        df = pd.DataFrame(columns=columns)
        df.loc[0] = [self.pmid, self.pmid_title, self.id, self.comment, self.disease_id, self.disease_label,
                     self.hgnc_id, self.gene_symbol, self.transcript_id, self.allele_1, self.allele_2, self.variant_comment,
                    self.age_of_onset, self.age_at_last_encounter, self.deceased,
                    self.sex, self.hpo]
        return df
```

```python
class Cohort_ET():

    def __init__(self, individuals: list):
        self.individuals = individuals

        # Join sets for the individuals
        self.hpo_set = set()
        for individual in individuals:
            self.hpo_set = self.hpo_set.union(individual.hpo_terms)
        print("[INFO] Found " + str(len(self.hpo_set)) + " HPO terms in total.")

        # Create table that shows for each of the joined HPO term if it is present in the three individuals
        hpo_table = []
        for hpo_term in self.hpo_set:
            hpo_row = []
            hpo_row.append(hpo_term._id)
            hpo_row.append(hpo_term._label)
            for individual in individuals:
                hpo_row.append(hpo_term in individual.hpo_terms)
            hpo_table.append(hpo_row)
        print(hpo_table)
        base_columns = ["HPO ID", "HPO Label"]
        individual_names = []
        for individual in individuals:
            individual_names.append(individual.id)
            #individual_names.append("Individual " + str(individual + 1))
        df = pd.DataFrame(hpo_table, columns=base_columns + individual_names)
        # Bring all rows with all columns TRUE to the top
        df = df.sort_values(by=individual_names, ascending=False)
        self.cohort_df = df

    def get_pandas_df(self):

        # Set base columns
        columns = ["PMID", "title", "individual_id", "comment", "disease_id", "disease_label"]
        columns += ["HGNC_id", "gene_symbol", "transcript", "allele_1", "allele_2", "variant.comment"]
        columns += ["age_of_onset", "age_at_last_encounter", "deceased", "sex", "HPO"]
        # Add columns for each HPO term
        for hpo_label in self.cohort_df["HPO Label"]:
            columns.append(hpo_label)
        df = pd.DataFrame(columns=columns)
        # Add first special row to the dataframe
        first_row = ['CURIE','str','str','optional','CURIE','str','CURIE','str','str','str','str','optional','age','age','yes/no/na','M:F:O:U','na']
        for hpo_id in self.cohort_df["HPO ID"]:
            first_row.append(hpo_id)
        df.loc[0] = first_row
        # Add the individuals to the dataframe
        for individual in self.individuals:
            individual_df = individual.get_pandas_df()
            # Add columns for each HPO term
            for hpo_term in self.hpo_set:
                individual_df[hpo_term._label] = hpo_term in individual.hpo_terms
            df = pd.concat([df, individual_df], ignore_index=True)
        # Replace logical TRUE with 'observed' and FALSE with 'na'
        df = df.replace(True, 'observed')
        df = df.replace(False, 'na')
        
        return df
```
